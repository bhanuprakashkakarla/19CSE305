```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import preprocessing as prep

data = pd.read_excel('19CSE305_LabData_Set3.1.xlsx', sheet_name = 'thyroid0387_UCI')
# finding the datatypes of each attribute
print(data.dtypes)
```

```python
data.replace('?', np.NaN, inplace=True)
data
```

```python
data.isnull().sum()
```

```python
one_hot_encoding_columns = ['referral source', 'Condition']
label_encoding_columns= ['sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick', 'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid', 'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH measured', 'T3 measured', 'TT4 measured', 'T4U measured', 'FTI measured', 'TBG measured']
```

```python
# entering the missing values with mode, median and mean
mode_columns = label_encoding_columns + one_hot_encoding_columns
for i in mode_columns:
    data[i] = data[i].fillna(data[i].mode()[0])

median_columns = ['TSH', 'T3', 'TT4', 'TT4', 'T4U', 'FTI', 'TBG']
for i in median_columns:
    data[i] = data[i].fillna(data[i].median())

data
```

```python
# label encoding schema
from sklearn import preprocessing as prepro
label_encoder = prepro.LabelEncoder()
for i in label_encoding_columns:
    data[i] = label_encoder.fit_transform(data[i])

# One hot encoding schema
referral_source = pd.get_dummies(data['referral source'])
condition=pd.get_dummies(data['Condition'])
data.drop(['referral source','Condition'],axis=1,inplace=True)
data=pd.concat([data,referral_source,condition],axis=1)
```

```python
# Normalizing all the attributes
scaler = prep.MinMaxScaler()
scaled_columns = ['TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG']
data_copy = data
scaled_df = scaler.fit_transform(data_copy[scaled_columns])
scaled_df = pd.DataFrame(scaled_df, columns = scaled_columns)
for i in scaled_columns:
    data[i] = scaled_df[i]
data
```

```python
# Checking for similarity between first two vectors
# Simple matching coefficient
observation_1 = data.iloc[0]
observation_2 = data.iloc[1]
count_num_smc = 0
count_den_smc = 0
for i in data.columns:
    if observation_1[i] == observation_2[i]:
        count_num_smc += 1
        count_den_smc += 1
    else:
        count_den_smc += 1
smc = count_num_smc / count_den_smc
print("The Simple matching coefficient (SMC) between two observation vectors is: ", smc, "\n")

# Jaccard coefficient
count_num_jc = 0
count_den_jc = 0
for i in data.columns:
    if observation_1[i] == observation_2[i] and observation_1[i] == 1:
        count_num_jc += 1
        count_den_jc += 1
    elif observation_1[i] != observation_2[i]:
        count_den_jc += 1
jc = count_num_jc / count_den_jc
print("The Jaccard coefficient (JC) between two observation vectors is: ", jc, "\n")
```

```python
# Cosine similarity measure
import math as mat
observation_dot_product = 0
for i in data.columns:
    observation_dot_product += observation_1[i] * observation_2[i]
length_observation_1 = 0
length_observation_2 = 0
for i in data.columns:
    length_observation_1 += observation_1[i] ** 2
length_observation_1 = mat.sqrt(length_observation_1)
for i in data.columns:
    length_observation_2 += observation_2[i] ** 2
length_observation_2 = mat.sqrt(length_observation_2)
cosine_similarity = observation_dot_product / (length_observation_1 * length_observation_2)
print("The cosine similarity between two observation vectors is: ", cosine_similarity, "\n")
```
